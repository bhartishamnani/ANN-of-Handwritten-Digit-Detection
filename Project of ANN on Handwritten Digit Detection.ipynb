# Importing Libraries

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the datasets

```python
data = keras.datasets.mnist
```

```python
(X_train, y_train),(X_test, y_test) = data.load_data()
```

```python
# Flattening the X_train & X_test
flat_X_train = X_train.reshape(len(X_train), 28*28)
flat_X_test = X_test.reshape(len(X_test), 28*28)
```

```python
flat_X_train.shape
```

# keras Sequential

.Units(Neurons): https://stackoverflow.com/questions/44747343/keras-input-explanation-input-shape-units-batch-size-dim-etc

.Activation : https://keras.io/api/layers/activations/

.Compiler : https://www.tutorialspoint.com/keras/keras_model_compilation.htm

```python
model = keras.Sequential([
                          keras.layers.Dense(units=10,
                                             input_shape=(784,),
                                             activation = 'sigmoid')

])

model.compile(
    optimizer='Adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(flat_X_train,y_train, epochs=5)
```

```python
new_X_train = X_train/255
new_X_test = X_test/255
```

```python
new_X_train[0][5]
```

```python
new_flat_X_train = new_X_train.reshape(len(new_X_train), 28*28)
new_flat_X_test = new_X_test.reshape(len(new_X_test), 28*28)
```

```python
model = keras.Sequential([
                          keras.layers.Dense(units=10,
                                             input_shape=(784,),
                                             activation = 'sigmoid')

])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(new_flat_X_train,y_train, epochs=5)
```

```python
model.evaluate(new_flat_X_test,y_test)
```

```python
y_pred = model.predict(new_flat_X_test)
```

```python
y_pred[0]
```

```python
np.argmax(y_pred[0])
```

```python
y_pred_label = [np.argmax(i) for i in y_pred]
```

```python
y_pred_label
```

```python
# Confusion Matrix

cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred_label)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Actual value')
plt.ylabel('Predicted value')
```

# Multiple Layers

```python
model = keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                          keras.layers.Dense(units=10,
                                             input_shape=(784,),
                                             activation = 'sigmoid'),

                         # Second Hidden Layers
                          keras.layers.Dense(units=100,
                                             activation = 'sigmoid')


])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train,y_train, epochs=5)
```
